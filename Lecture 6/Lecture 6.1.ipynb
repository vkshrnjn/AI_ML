```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
xpoints = np.array([0,5])
ypoints = np.array([0,200])
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(xpoints, ypoints, 'x')
plt.show()
```

```python
plt.plot(xpoints, ypoints,marker='D')
plt.show()
```

```python
help(plt.plot)
```

```python
xpoints = np.array([1,2,4,6])
```

```python
ypoints = np.array([3,7,4,10])
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

```python
plt.plot(xpoints,ypoints,'*')
plt.show()
```

```python
plt.plot(xpoints,ypoints,linestyle="dotted")
plt.show()
```

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("this is a demo graph")
plt.show()
```

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("this is a demo graph")
plt.plot(xpoints,ypoints)
plt.show()
```

```python
font1 = {"family":'serif',"color":'red', "size":20}
font2 = {"family":'serif',"color":'blue', "size":10}
```

```python
plt.title("this is a title", fontdict = font1)
plt.xlabel("this is x-axis", fontdict = font2)
plt.ylabel("this is y-axis", fontdict = font2)
plt.plot(xpoints,ypoints)
plt.show()
```

```python
df_by_season = bikes.groupby("season")
```

```python
df_seasons = df_by_season.cnt.sum().sort_values()
```

```python
plt.ticklabel_format(style="plain")
plt.plot(bikes.season.unique(),df_by_season.cnt.sum())
```

```python
fig, ax = plt.subplots()
index = bikes.season.unique()
bar_width = 0.5

values = df_by_season.cnt.sum()
ax.bar(index, values, bar_width)
```

```python
fig, ax = plt.subplots()
ax.scatter(bikes.casual, bikes.registered)
plt.show()
```

```python
fig, ax = plt.subplots()
values = df_by_season.cnt.sum()
labels = ["s1","s2","s3","s4"]

ax.pie(values, labels=labels)
plt.show()
```
